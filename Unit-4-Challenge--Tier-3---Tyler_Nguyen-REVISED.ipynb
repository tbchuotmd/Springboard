````python
# Subset entries from most recent year (2024), sort & group by Borough, then calculate average price of properties in each Borough; save to avgpb24
avgpb24 = props_tidy_nona.query('year == 2024').sort_values('Borough').groupby('Borough').agg({'Price':'mean'})
avgpb24.head()
````

````python
# Subset entries from two decades prior (2004), sort & group by Borough, then calculate average price of properties in each Borough; save to avgpb04
avgpb04 = props_tidy_nona.query('year == 2004').sort_values('Borough').groupby('Borough').agg({'Price':'mean'})
avgpb04.head()
````

````python
# Join average prices for 2004 with that for 2024, saving as pb20ygap.
pb20ygap = avgpb04.merge(avgpb24,on='Borough',suffixes=('-AVG_2004','-AVG_2024'))
pb20ygap.head()
````

````python
# Calculate Price Ratios (2024 : 2004) and append as PR column
pb20ygap['PR'] = pb20ygap['Price-AVG_2024'] / pb20ygap['Price-AVG_2004']
pb20ygap.head()
````

````python
# Sort by price ratio in descending order to see the highest price increases listed first, assign to pb20yg_sorted.
pb20yg_sorted = pb20ygap.sort_values('PR', ascending=False)
pb20yg_sorted.head()
````

````python
# Create list of boroughs ranked by average price ratio (2024 : 2004) in descending order, assign to boroughs_byPRdesc
boroughs_byPRdesc = list(pb20yg_sorted.index)
boroughs_byPRdesc
````

### REVISION BEGINS HERE:

Upon reading about the geography, it became apparent that several items under the "Borough" field don't actually constitute a London borough.  In fact, the original dataset included broader regions encompassing multiple boroughs (EAST OF ENGLAND, LONDON, SOUTH EAST, etc.) along with “City of London”, which is technically not a borough despite being within London.

The following details how the data was further cleaned to include only the boroughs of London, starting back from ```props_tidy_nona```, which is the dataframe containing everything in the proper data format with null values dropped, and after transposing/melting.

````python
# Subset entries containing 'london' in their name; these are not actually boroughs.
london_ovw = props_tidy_nona[props_tidy_nona['Borough'].str.contains('london',case=False)]
````

````python
london_ovw['Borough'].value_counts()
````

````python
# Read in list of boroughs from wikipedia for comparison.
wikiurl = "https://en.wikipedia.org/wiki/List_of_London_boroughs"
bors = pd.read_html(wikiurl)[0]
bors['Borough']
````

````python
# Remove [notes] from borough names using REGEX patterns
import re
pattern = r'\[.*\]'
bors['Borough'] = bors['Borough'].apply(lambda x: re.sub(pattern, '', x))
bors['Borough']
````

````python
# Remove entries with names containing 'london' from props_tidy_nona, save as ptidy_rev (rev for revision)
ptidy_rev = props_tidy_nona[~props_tidy_nona['Borough'].isin(london_ovw['Borough'])]
ptidy_rev['Borough'].drop_duplicates()
````

````python
# Preview of removing entries with names that are in all caps; these are broader categories encompassing multiple buroughs
ptidy_rev[~ptidy_rev['Borough'].str.isupper()]['Borough'].drop_duplicates()
````

````python
# Combine removing 'london' and all-caps entries, this time altering the actual dataframe
ptidy_rev = ptidy_rev[~ptidy_rev['Borough'].str.isupper() & ~ptidy_rev['Borough'].str.contains('England')]
````

````python
# Sort by borough, then date
ptdf = ptidy_rev.sort_values(['Borough','Date'])
ptdf.head()
````
