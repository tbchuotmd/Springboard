```python
data = np.round(np.random.normal(68.9, 10, 6000),10)
len(data)
np.mean(data)
np.median(data)
data2 = np.round(np.random.normal(3.568, 1234, 6000),10)
print(f"CORRELATION COEFFICIENT:\n{np.corrcoef(data,data2)}")
tbl = np.column_stack((data,data2))

print(f"\nTABLE:\n{tbl}\n")

plt.hist(data, bins=100)
plt.show()
plt.cla()
plt.hist(data2, bins=100)
plt.show()
```
